````python
import os 
from dotenv import load_dotenv
from IPython.display import display, Markdown
from openai import OpenAI
from IPython.utils.capture import capture_output
import base64
````

````python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
````

````python
my_self="""Full-Stack Software Engineer with 4 years industry experience specializing in Node.js, React, Next.js, Go and backend architecture. Experienced in building scalable applications, AI integrations, and REST APIs. Strong background in MongoDB, PostgreSQL and Frontend 
environments. Looking for a Software Engineer opportunity to build high-performance systems."""
````

````python
# ATS-friendly resume prompt
ats_resume_prompt = f"""Create an ATS-friendly resume for this candidate:

CANDIDATE INFO:
{my_self}

STRICT ATS REQUIREMENTS:
1. NO tables, columns, graphics, or images
2. NO headers/footers
3. Use standard section headings only: CONTACT, SUMMARY, TECHNICAL SKILLS, WORK EXPERIENCE, EDUCATION, CERTIFICATIONS
4. Use simple bullet points (• or -)
5. NO special characters, emojis, or icons
6. Use standard fonts (implied in plain text)
7. Include keywords from job descriptions (Node.js, React, Next.js, Go, MongoDB, PostgreSQL, REST APIs, AI)
8. Use standard date format: Month Year - Month Year
9. Quantify achievements with numbers where possible
10. NO columns or multi-column layouts

FORMAT AS PLAIN TEXT WITH THIS STRUCTURE:

CONTACT
[Name: Full-Stack Software Engineer]
[Email: placeholder@email.com]
[Phone: (555) 123-4567]
[LinkedIn: linkedin.com/in/placeholder]
[GitHub: github.com/placeholder]

SUMMARY
[2-3 sentences with years of experience, key technologies, and value proposition]

TECHNICAL SKILLS
• Backend: [list]
• Frontend: [list]
• Databases: [list]
• Tools & Others: [list]

WORK EXPERIENCE
Full-Stack Software Engineer | [Company Name] | [City, State]
[Month Year] - Present
• [Achievement with number/metric]
• [Responsibility with keyword]
• [Technical implementation detail]

[Previous Role] | [Company] | [Location]
[Month Year] - [Month Year]
• [Achievement]
• [Responsibility]

EDUCATION
[Bachelor's in Computer Science] | [University Name] | [Location]
[Year Graduated]

CERTIFICATIONS
• [Relevant Certification]
• [Relevant Certification]

Make every bullet point keyword-rich for ATS parsing. Use numbers and metrics. Keep it single-column, plain text format."""
````

````python
system_promt="You are an ATS resume expert. Create simple, keyword-rich, plain text resumes that pass automated screening systems."
````

````python
user_promt=f"{ats_resume_prompt}"
````

````python
from openai import OpenAI
openai = OpenAI()
response = openai.chat.completions.create(
    model = "gpt-5-nano",
    messages = [
         {"role": "system", "content": system_promt},
        {"role": "user", "content": user_promt}
    ]
)
#response.choices[0].message.content
ats_resume = response.choices[0].message.content
````

````python
print("="*60)
print("📄 ATS-FRIENDLY RESUME")
print("="*60)
print("\n")
display(Markdown(f"```\n{ats_resume}\n```"))
````

````python
print("\n" + "="*60)
print("📥 DOWNLOAD OPTIONS")
print("="*60)
````

````python
b64 = base64.b64encode(ats_resume.encode()).decode()
download_link_txt = f'<a href="data:text/plain;base64,{b64}" download="ats_resume.txt">📥 Download ATS Resume (TXT - Best for ATS)</a>'
display(Markdown(download_link_txt))
````
